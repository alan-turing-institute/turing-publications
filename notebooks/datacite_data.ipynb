# Brief look at DataCite data

Looking at publishers and year published.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/datacite_extracted_data.csv")  # Adjust the path as needed
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Publisher', order=df['Publisher'].value_counts().index)
plt.title('Count of Publications by Publisher')
plt.xlabel('Count')
plt.ylabel('Publisher')
plt.show()
```

```python
# Group the data by 'Publisher' and count the number of publications
publisher_counts = df.groupby('Publisher').size().reset_index(name='Number of Publications')

# Sort the data by the number of publications in descending order
publisher_counts = publisher_counts.sort_values('Number of Publications', ascending=False)

# Display the table
publisher_counts
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Year Published'], kde=False)
plt.title('Distribution of Publication Dates')
plt.xlabel('Year Published')
plt.ylabel('Count')
plt.show()
```
